# Extract Data from Default Noise File

- 8 Layers - each has 1 histogram
- 83 bins in each histogram 

```c++
TFile *f=new TFile("elecNoise_ecalBarrelFCCee_50Ohm_traces1_4shieldWidth.root");
```

```c++
f->ls();
```

TH1F::GetBinContent()
- bin = 0 is the underflow bin 
- bin = 1 is the first bin with low-edge xlow included 
- bin = nbins is the last bin with upper-edge xup excluded 
- bin = nbins + 1 is the overflow bin

## Layer 1 

```c++
TH1F *h1=(TH1F*)f->Get("h_elecNoise_fcc_1");
h1->Draw();
c1->Draw();

// bin = 0 is underflow therefore start from i=1
int nbins = 84;

// write to text file 
ofstream MyFile("DefaultNoiseValues_Layer1.txt");

for(Int_t i=1; i<nbins; i++){
    
    double noise = h1->GetBinContent(i);
    MyFile << noise << endl;
  
}

MyFile.close();
```

```c++
// write to text file 
ofstream MyFile("etaValues.txt");

for(Int_t i=1; i<nbins; i++){
    
    double eta = h1->GetXaxis()->GetBinCenter(i);
    MyFile << eta << endl;
  
}

MyFile.close();
```
